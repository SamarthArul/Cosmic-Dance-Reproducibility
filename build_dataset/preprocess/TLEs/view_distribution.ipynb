### Before cleaning

```python
CDF_CSV = "/mnt/Storage/OUTPUTs/Starlink/OUTPUTs/pre_process_TLEs/before_cleanup.csv"

df_tles = get_merged_TLEs_from_all_CSVs(TLE_DIR_CSV)


# Save the CDF in CSV
x, y = get_CDF(df_tles[TLE.ALTITUDE_KM])
pd.DataFrame.from_dict(
    {
        "X": x, "Y": y
    }
).to_csv(CDF_CSV)


# Plot histogram
plt.hist(df_tles[TLE.ALTITUDE_KM])
plt.xlabel("Altitude (km)")
plt.ylabel("# Count")
plt.grid()
plt.tight_layout()
plt.show()
```

### After cleaning

```python
CDF_CSV = "/mnt/Storage/OUTPUTs/Starlink/OUTPUTs/pre_process_TLEs/after_cleanup.csv"

# Save the CDF in CSV
df_tles = get_merged_TLEs_from_all_CSVs(TLE_DIR_CSV)
x, y = get_CDF(df_tles[TLE.ALTITUDE_KM])
pd.DataFrame.from_dict(
    {
        "X": x, "Y": y
    }
).to_csv(CDF_CSV)

# Plot histogram
plt.hist(df_tles[TLE.ALTITUDE_KM])
plt.xlabel("Altitude (km)")
plt.ylabel("# Count")
plt.grid()
plt.tight_layout()
plt.show()
```

### Compare the distributions

```python
df_clean = read_CSV("/mnt/Storage/OUTPUTs/Starlink/OUTPUTs/pre_process_TLEs/after_cleanup.csv")
df_raw = read_CSV("/mnt/Storage/OUTPUTs/Starlink/OUTPUTs/pre_process_TLEs/before_cleanup.csv")
```

```python
# Plotting
SIZE = 17

plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams.update({'font.size': 25})

# CDF before cleaning up
plt.scatter(df_raw["X"], df_raw["Y"], s=SIZE, c='r')

plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
plt.xticks([1000, 10000, 20000, 30000, 40000])
plt.xlim(100, 40000)

plt.xlabel("Altitude (km)")
plt.ylabel("CDF")
plt.grid()
plt.tight_layout()
# plt.savefig("/home/suvam/Projects/CosmicDance/figures/pre_process/TLEs/before_cleanup.png")
plt.plot()
```

```python
plt.rcParams["figure.figsize"] = (7, 5)
plt.rcParams.update({'font.size': 25})

# CDF after cleaning up
plt.scatter(df_clean["X"], df_clean["Y"], s=SIZE, c='r')

plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
plt.xticks([100, 200, 300, 400, 550])
plt.xlim(100, 615)

plt.xlabel("Altitude (km)")
plt.ylabel("CDF")

plt.grid()
plt.tight_layout()
# plt.savefig("/home/suvam/Projects/CosmicDance/figures/pre_process/TLEs/after_cleanup.png")
plt.plot()
```
