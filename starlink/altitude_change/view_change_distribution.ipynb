```python
from cosmic_dance.dst_index import *
from cosmic_dance.io import *
from cosmic_dance.stack_plots import *
from cosmic_dance.stats import *
from cosmic_dance.TLEs import *
```

```python
plt.rcParams["figure.figsize"] = (10, 7)
plt.rcParams.update({'font.size': 30})
SIZE = 7

def altitude_change_CDF(df: pd.DataFrame):
    # Altitude change
    x, y = cdf(df["KM_after_DAY_1"])
    plt.scatter(
        x, y,
        color='g', s=SIZE, marker="o",
        label='After 1 days'
    )

    x, y = cdf(df["KM_after_DAY_5"])
    plt.scatter(
        x, y,
        color='b', s=SIZE, marker="^",
        label='After 5 days'
    )

    x, y = cdf(df["KM_after_DAY_10"])
    plt.scatter(
        x, y,
        color='r', s=SIZE, marker="D",
        label='After 10 days'
    )

    plt.xlabel('Absolute altitude difference (km)')
    plt.ylabel('CDF')
    plt.xlim(0, 150)
    plt.ylim(0.98, 1.001)
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.title('Altitude change')
    plt.show()

def drag_change_CDF(df: pd.DataFrame):
    # Drag change
    x, y = cdf(df["DRAG_after_DAY_1"])
    plt.scatter(
        x, y,
        color='g', s=SIZE, marker="o",
        label='After 1 days'
    )

    x, y = cdf(df["DRAG_after_DAY_5"])
    plt.scatter(
        x, y,
        color='b', s=SIZE, marker="^",
        label='After 5 days'
    )

    x, y = cdf(df["DRAG_after_DAY_10"])
    plt.scatter(
        x, y,
        color='r', s=SIZE, marker="D",
        label='After 10 days'
    )

    plt.xlabel('Absolute drag')
    plt.ylabel('CDF')
    plt.ylim(0.98, 1.001)
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.title('Drag')
    plt.show()
```

### Altitude change in normal days

```python
df_quiet_days = read_CSV("../../artifacts/OUTPUT/Starlink/measurement/maximum_altitude_change/quiet_day_after_1_5_10.csv")

altitude_change_CDF(df_quiet_days)
drag_change_CDF(df_quiet_days)
```

### Altitude change after solar events

```python
df_event_days = read_CSV("../../artifacts/OUTPUT/Starlink/measurement/maximum_altitude_change/event_day_after_1_5_10.csv")

altitude_change_CDF(df_event_days)
drag_change_CDF(df_event_days)
```
